```python
#Exemplo de query
hs.query('price > 5000000 & bedrooms > 5')
```

```python
hs['mq_interna'] = hs['sqft_living'].apply(lambda x: x/10.764)
hs['mq_lote'] = hs['sqft_lot'].apply(lambda x: x/10.764)
hs['mq_piso_superior'] = hs['sqft_above'].apply(lambda x: x/10.764)
hs['mq_porao'] = hs['sqft_basement'].apply(lambda x: x/10.764)

#mq_porao.sort_values().head()
```

```python
#Concatena a latitude e a longitude, convertendo ela em string
hs['latlon'] = hs[['lat','long']].astype(str).agg(','.join, axis=1)
```

```python
#Entendendo a variavel grade
#Quanto maior o grau maior o preço, geralmente casas de pessoas com poder aquisitivo maior
```

```python
hs.query('grade > 6')[['condition','grade','price','lat','long','latlon']].sort_values(by='price',ascending=False).head()
```

```python
hs.query('grade <= 5')[['condition','grade','price','lat','long','latlon']].sort_values(by='price',ascending=False).head()
```

```python
#Ententendo a variavel condition:

#condition é basicamente a condição da casa 
#Baixo = casa velha/detonada 
#Alto = casa nova/bonitona
```

```python
hs.query('condition >= 4')[['condition','grade','price','lat','long','latlon']].sort_values(by='price',ascending=False).head()
```
